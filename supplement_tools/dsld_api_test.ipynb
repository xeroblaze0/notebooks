```typescript
async function get_data(user_search_query){
    try {
        const url = "https://api.ods.od.nih.gov/dsld/v9/"
        // const search_location = "ingredient-groups/"
        const search_location = "browse-products/"
        const search_method = "?method=by_letter"
        const search_query_tag = "&term="
        
        // let user_search_query = "K"
        let encoded_search_query = encodeURIComponent(user_search_query)
        
        let site = url+search_location+search_method+search_query_tag+encoded_search_query

        const res = await fetch(site);

        if (!res.ok){
            throw new Error('Request failed.');
        }

        const data = await res.json();

        //Do something with the decoded json data
        // console.log(data)
        return data.hits;

    } catch (e){
        console.log('There was an error.');
    }
}
```

```typescript

async function collect_data(){
    const data = await get_data('C');
    console.log(data);
}

collect_data()
```

```typescript
async function sort_data(){
    const data = await get_data('C')

    let arr = [];

    for (var i in data){
        var obj = {};
        obj = data[i]._source
        arr.push(obj);
    }
    arr.sort((a, b) => a.groupName > b.groupName ? 1 : -1);

    // console.log(arr);
    return arr;
}

const stuff = await sort_data()
console.log(stuff)
```
