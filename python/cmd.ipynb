```python
# https://gist.github.com/mikkkee/73cf05969a97e806dea6?permalink_comment_id=4473182#gistcomment-4473182

import time

def spinning_cursor():

  speed = 0.075     # lower = faster spin
  duration = 1.0   # length of time spinning
  
  start_time = time.time()

  while time.time() - start_time < duration:
    for cursor in '\\|/-':
      time.sleep(speed)
      print(f"\r{cursor}", end="", flush=True)
    pass

spinning_cursor()
```
